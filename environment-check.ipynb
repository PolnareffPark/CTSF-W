```python
import platform, sys, torch, numpy as np, os

print("OS:", platform.platform())
print("Python:", sys.version)
print("PyTorch:", torch.__version__, "CUDA:", torch.version.cuda, "cuDNN:", torch.backends.cudnn.version())
print("NumPy:", np.__version__)
#print("BLAS:", np.__config__.get_info('blas_opt_info'))
print("GPU:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")
print("Env OMP/MKL:", os.getenv("OMP_NUM_THREADS"), os.getenv("MKL_NUM_THREADS"))
```
